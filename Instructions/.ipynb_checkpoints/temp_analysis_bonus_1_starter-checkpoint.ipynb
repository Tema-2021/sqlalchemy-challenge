### Compare June and December data across all years 

```python
from scipy import stats
hawaii_measurements=pd.read_csv('Resources/hawaii_measurements.csv')
hawaii_stations=pd.read_csv('Resources/hawaii_stations.csv')
```

```python
# Identify the average temperature for June
june_data = hawaii_measurements[hawaii_measurements['date'].str.slice(start=5, stop=7)=='06'].groupby(hawaii_measurements['station']).mean()
```

```python
# Create collections of temperature data
#Display output
june_data['tobs']
```

```python
june_data.describe()
```

```python
# Average Temp for June was 74.853510
```

```python
# Identify the average temperature for December
dec_data = hawaii_measurements[hawaii_measurements['date'].str.slice(start=5, stop=7)=='12'].groupby(hawaii_measurements['station']).mean()
```

```python
# Create collections of temperature data
#Display output
dec_data['tobs']
```

```python
dec_data.describe()
```

```python
# Average tempersture for December was 71.219070
```

```python
# Run paired t-test
stats.ttest_ind(june_data['tobs'], dec_data['tobs'], equal_var=False)
```

### Analysis

```python
# Average Temp for June was 74.853510
# Average tempersture for December was 71.219070
# The null hypothesis is being rejected and records are statistically significant as pvalue is 0.00037
```
